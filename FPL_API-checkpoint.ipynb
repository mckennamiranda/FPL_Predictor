```python
threat_per_goal = []

for b,m in zip(threat,goals):
    try:
        threat_per_goal.append(float(b)/float(m))
    except ZeroDivisionError:
        threat_per_goal.append(0)
threat_per_goal
```

```python
bps_per_bonus = []

for b,m in zip(bps,bonus):
    try:
        bps_per_bonus.append(b/m)
    except ZeroDivisionError:
        bps_per_bonus.append(0)
bps_per_bonus
```

```python
creativity_per_assist = []

for b,m in zip(creativity,assists):
    try:
        creativity_per_assist.append(float(b)/float(m))
    except ZeroDivisionError:
        creativity_per_assist.append(0)
creativity_per_assist
```

```python
ict_index_per_total_points = []

for b,m in zip(ict_index, total_points):
    try:
        ict_index_per_total_points.append(float(b)/float(m))
    except ZeroDivisionError:
        ict_index_per_total_points.append(0)
ict_index_per_total_points
```

# Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python


Stats_df = Stats_df.dropna()
Stats_df = Stats_df.reset_index(drop=True)
Stats_df
```

```python

        
        
Stats_filtered = Stats_df[Stats_df['minutes'] != 0]
Stats_filtered
```

```python
# Assign the data to X and y

X = Stats_filtered.iloc[:,3:33]
y = Stats_filtered["total_points"].values.reshape(-1, 1)
print(X.shape, y.shape)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
model = LinearRegression()
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
```
